```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))
print("\n3. Array of ones:\n", ones_array)
print("\nArray of zeroes:\n", zeros_array)
```

```python
random_6x3 = np.random.rand(6, 3)
transpose_array = random_6x3.T
print("\n4. Random 6x3 array:\n", random_6x3)
print("\nTransposed array (3x6):\n", transpose_array)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)
print("\n5. Matrix A (n x p):\n", A)
print("\nMatrix B (p x m):\n", B)
print("\nDot product (n x m):\n", dot_product)
```
